```python
import pandas as pd

data_o = pd.read_csv('temp_popular_minmax_part.csv')
data_o.head()
```

```python
genres = ['hiphop', 'funk', 'disco', 'electronic', 'blues', 'world', 'movie', 'classical']
countries = ['Europe', 'Africa', 'Americas', 'Asia', 'Oceania']

diction_gc = {}
for genre in genres:
    diction_gc[genre] = []
    for country in countries:
            data_i = data_o[data_o['genre'] == genre]
            data_i = data_i[data_i['continent'] == country]
            pop = int(data_i['part'])
            diction_gc[genre].extend([1] * pop)
            diction_gc[genre].extend([0.2] * (5-pop))

```

```python
for key, value in diction_gc.items():
    print(key)
    print(value)
```
